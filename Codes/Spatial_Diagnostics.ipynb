# Standardize the variables

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# 1) Variables to standardize -------------------------------------------------
cols_to_keep = [
    "Forest", "Shrub", "Urban", "Crop", "Water", "Other",
    "Mn_P_Wn", "Mx_WD_W", "Mx_DD_W",
    "Mn_T_Sp", "Mx_DD_Sp",
    "Mn_T_Sm", "Mx_WD_A", "Mx_DD_A",
    "gdp", "pop_dns", "incidnc"
]

gdf_scaled = gdf.copy()
gdf_scaled = gdf_scaled.loc[~gdf_scaled.geometry.is_empty].dropna(subset=['geometry'])

# 2) Crete the scaler
scaler = StandardScaler()
# 3) Apply the scaler to cols_to_keep
gdf_scaled[cols_to_keep] = scaler.fit_transform(gdf_scaled[cols_to_keep])
```

# Compute Weighted metric W

```python
from libpysal.weights import DistanceBand
import numpy as np
gdf_scaled_proj = gdf_scaled.to_crs(32633)
coords = np.column_stack([
    gdf_scaled_proj.geometry.centroid.x,
    gdf_scaled_proj.geometry.centroid.y
])
max_distance = 150_000  # 150 km
# DistanceBand excludes self-neighbors by default; binary=True matches style="B"
# silence_warnings=True is akin to zero.policy=TRUE (it won’t error on isolates)
W = DistanceBand(
    coords,
    threshold=max_distance,
    binary=True,
    silence_warnings=True  # analogous to zero.policy handling in R
)

# --- 4) Get coordinates if you need them (like coordinates()) ---
# (Not required for DistanceBand.from_dataframe, but included for parity with R)

# --- 5) Convert weights to a matrix (like listw2mat) ---
# W.sparse is a scipy CSR matrix in the order W.id_order.
W_matrix = W.sparse.toarray()
```

# Spatial lag of Y variable

```python
y_ = gdf_scaled_proj['incidnc'].to_numpy(dtype=float)
# Spatial lag: W * y
y_lag = W_matrix @ y_
```
